```python
def path_to_vm(p, clusters=10):
    p = [p[x+3] - p[x] for x in range(len(p) - 3)]
    p = np.reshape(p, (-1, 3))
    norm = np.sqrt((p * p).sum(axis=1))
    p = p / norm.reshape(-1,1)
    vmf = spherecluster.VonMisesFisherMixture(n_clusters=clusters, posterior_type='soft')
    labels = vmf.fit_predict(p)
    dirs = vmf.cluster_centers_
    conc = vmf.concentrations_
    values = [(dirs[i].tolist(),conc[i]) for i in labels]
    s = simplifyList(values)
    return s

def path_to_movmf(p, clusters=10):
    p = [p[x+3] - p[x] for x in range(len(p) - 3)]
    p = np.reshape(p, (-1, 3))
    norm = np.sqrt((p * p).sum(axis=1))
    p = p / norm.reshape(-1,1)
    vmf = spherecluster.VonMisesFisherMixture(n_clusters=clusters, posterior_type='soft')
    vmf.fit(p)
    dirs = vmf.cluster_centers_ #direction of each cluster on the unit sphere
    kappas = vmf.concentrations_ #concentration parameter (higher = more concentrated)
    alphas = vmf.weights_ #weight of each cluster
    out = []
    for i in range(clusters):
        #surely there's a better way, but this'll work
        a = []
        a.append(alphas[i])
        a.append(kappas[i])
        a.extend(dirs[i].tolist())
        out.append(a)
    return out
    
def simplifyList(v):
    out = []
    for dirs,conc in v:
        out.extend(dirs)
        out.append(conc)
    return out
```
